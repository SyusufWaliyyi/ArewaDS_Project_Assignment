## __Import the required modules__
In this section, the modules needed to be imported for the project will be imported
1. Matplotlib will be imported as plt
2. the module random_walk created will be imported

```python
import matplotlib.pyplot as plt
from random_walk import RandomWalk
```

## __Plotting the Random Walk__
To visualize the random walk, we can use the `matplotlib` library imported. This will enable us to plot the random walk points shown below.

```python
rW = RandomWalk()

rW.fill_walk()
plt.plot(rW.x_values, rW.y_values)
plt.show()
```

## __Generating Multiple Random Walks__
To avoid running the program repeatedly while loop will be used to wrap the code as shown below
The code prompt the user to enter either y or n. y stands for yes while n stand for no. y will keep the code running and n to stop plotting.

```python
while True:
    plt.plot(rW.x_values, rW.y_values)
    plt.show()
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

## __Styling the Walk__
To style this walk, the point is to emphasize on the most important aspect of the walk which include
    1. Where the walk begin, where it ended and the path taken
    2. Tick marks and labels 
at the end a clear presentation of the walk will be shown

### _Colouring the Points_ 
Color argument is going to be used to show the order of the points in the walk and then remove the black outline. 
c argument list will be pass containing the positions in the walk. Remember the points started from 0 to 5000

```python
while True:
    point_numbers = list(range(rW.num_points))
    plt.plot(rW.x_values, rW.y_values, linewidth=1, c='red', cmap=plt.cm.Blues)
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

## __Plotting the Starting and Ending Points__
To visualize the path taken by the robot, we need to plot the starting and ending points. The ending points will be make larger than the starting points and will be coloured differently.

```python
while True:

    point_numbers = list(range(rW.num_points))
    plt.plot(rW.x_values, rW.y_values, c='blue', linewidth=1)
# Emphasize the first and last points.
    plt.plot(0, 0, c='red', marker = 'X', markersize = 10, linewidth= 30)
    plt.plot(rW.x_values[-1], rW.y_values[-1], c='green', marker= 'o', markersize = 10, linewidth=30)
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

### __Cleaning Up the Axes__
Here the axes are going to be removed from the plot, remember they do not distruct us from the path of each walk

```python
while True:
    plt.scatter(rW.x_values[-1], rW.y_values[-1], c='red', edgecolors='none', s=100)
    
    # Remove the axes
    plt.axes().get_xaxis().set_visible(False)
    plt.axes().get_yaxis().set_visible(False)
    plt.title('Cleaning up the axes')
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

## __Adding Plot Points__
Its possible to increase the number of points to generate more data points. This will increase the value of num_points when the RandomWalk instance is created and similarly adjust the size of each dot when drawing the plot points as shown below

```python
while True:
# Make a random walk, and plot the points.
    rW = RandomWalk(5000)
    rW.fill_walk()
# Plot the points, and show the plot.
    point_numbers = list(range(rW.num_points))
    plt.plot(rW.x_values, rW.y_values, c='green', linewidth=3)
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break 
```

The plot above creates a random walk of 5000 points and plotted each point at size s = 1. The resulting plots is wispy and looks like a cloud

## __Altering the Size to Fill the Screen__

The size of the plot can be adjusted to fill the entire screen. The code below will set the figure size to be 10 inches by 6 inches. The resulting plots will be larger and more visually appealing.

```python
while True:
    # Set the size of the plotting window.
    plt.figure(figsize=(10, 6))
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```
