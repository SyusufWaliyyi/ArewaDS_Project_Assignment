## _Calculating Data Automatically_

Entering data manually can be time consuming and is prone to making a lot of mistakes. This can be avoided by using loops to iterate over a set of data.
Example range can be used to enter a large range of a data set and then iterate over that data set using a loop that returns a list of integers in the range specified by the range parameter.This range can be  square of the range parameter, even, odd, prime numbers over the range parameter and so on. Consider the example below.

```python
x_values = list(range(1, 1001)) # range from 1 to 1001
y_values = [x**3 for x in x_values] # squares of x_values

#print(x_values) # prints the x_values
#print(y_values) # prints the y_values
plt.scatter(x_values, y_values, s=40)
plt.show()
```

what does the data above mean? 
x_values: Contain a list values from 1 to 1000
y_values: This was generated by looping through the x values(for x in x_values) using list comprehension and then taking square of each x_value (x**2) and assigning ot y values. 

The graph can be styled the same way as the ones above

```python
x_values = list(range(1, 1001)) # range from 1 to 1001
y_values = [x**2 for x in x_values] # squares of x_values

#print(x_values) # prints the x_values
#print(y_values) # prints the y_values
plt.scatter(x_values, y_values, s=40)
plt.title('Square Numbers', fontsize=20)
plt.xlabel('Values', fontsize=20)
plt.ylabel('Square of Values', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.axis([0, 1100, 0, 1100000]) # Set the range for each axis.
plt.grid(True)
plt.show()
```

plt.axis is the added styling here. This is called due to the large data range we have. X axis was set from 1 to 1100 while Y axis was set from 1 to 1100000 for easy readability.

## _Removing Outlines from Data Points_

matplotlib grant you a previledge of colour individually. Meaning a certain data points can be coloured entirely different from the other data points

```python
plt.scatter(x_values, y_values, edgecolor='none', s=40) # This will output similar graph of solid blue colour
plt.show()
```

## _Defining Custom Colors_

The colour of a point in a scatter plot can be change in two ways
1. By adding a code to the plot donated by c and assign a 
   colour name or the colour initials in a string format. Example: let plot using red (r), green (g) and black (k)

```python
plt.scatter(x_values, y_values, color= 'g',  edgecolor='none', s=40) # This will output similar graph of solid blue colour
plt.show()
```

```python
plt.scatter(x_values, y_values, c= 'r',  edgecolor='none', s=40) # This will output similar graph of solid blue colour
plt.show()
```

```python
plt.scatter(x_values, y_values, c= 'k',  edgecolor='none', s=40) # This will output similar graph of solid blue colour
plt.show()
```

2. Using RGB colour system: To define a colour pass the
   variables defined as c a tuple with three decimal values (1 one each for red, green and blue), using values between 0 and 1. Example c = (0,0,0.3) will a dark blue colour while c = (0,0,0.8) will produce light blue plot.

```python
plt.scatter(x_values, y_values, color= (1,0,0),  edgecolor='none', s=40) # This will output similar graph of solid blue colour
plt.show()
```

```python
plt.scatter(x_values, y_values, color= (0.5,0.5,0.5),  edgecolor='none', s=40) # This will output similar graph of solid blue colour
plt.show()
```

## _Using a Colourmap_

This a gradient system of colouring a plot whose concentration increases from start to end. It is mostly useful used when a user wants to emphasize a portion of the data set. One important thing to note is that there is need to specify how the ppylot should assign the colour to each data point. Let see the example below for using colourmap to plot red and blue colour

```python
plt.scatter(x_values, y_values, c=x_values, cmap=plt.cm.Reds,
edgecolor='none', s=40)
plt.show()
```
